```python
# Import BeautifulSoup
from bs4 import BeautifulSoup as bs
response = requests.get("https://www.osti.gov/dataexplorer/api/v1/records")
html = response.content

bs_content = bs(html, "html")  #since the url seems to be an XML file, so I use "lxml' parsing here"
```
